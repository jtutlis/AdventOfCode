```python
f = open("data.txt", "r")
data = f.readline().split(',')
intcode = [int(x) for x in data]

intcode[1] = 12
intcode[2] = 2
```

```python
i = 0
while intcode[i] != 99:
    instruction = intcode[i]
    pos1 = intcode[intcode[i+1]]
    pos2 = intcode[intcode[i+2]]
    if instruction == 1:
        val = pos1 + pos2
    else:
        val = pos1 * pos2
    intcode[intcode[i+3]] = val

    i += 4
```
