```python
from timeit import default_timer as timer

f = open("data.txt", "r")
data = f.readlines()
data = [x.strip('\n') for x in data]
data = [(x[0], int(x[1:])) for x in data]
```

```python
def part1(instructions):
    currDirection = 0
    up = 0 # northward
    side = 0 # eastward

    for action, value in instructions:
        if action == 'E':
            side += value
        elif action == 'W':
            side -= value
        elif action == 'N':
            up += value
        elif action == 'S':
            up -= value
        elif action == 'R':
            currDirection = (currDirection + value) % 360
        elif action == 'L':
            currDirection = (currDirection - value) % 360
        elif action == 'F':
            if currDirection == 0:
                side += value
            elif currDirection == 180:
                side -= value
            elif currDirection == 270:
                up += value
            elif currDirection == 90:
                up -= value
        else:
            print(action)
    return abs(up) + abs(side)
        

    
```

```python
part1(data)
```

```python
def part2(instructions):
    waypointUp = 1 # northward
    waypointSide = 10 # eastward

    shipUp = 0 # northward
    shipSide = 0 # eastward

    for action, value in instructions:
        if action == 'E':
            waypointSide += value
        elif action == 'W':
            waypointSide -= value
        elif action == 'N':
            waypointUp += value
        elif action == 'S':
            waypointUp -= value
        elif action == 'R':
            if value == 90:
                waypointUp, waypointSide = -waypointSide, waypointUp
            elif value == 180:
                waypointUp, waypointSide = -waypointUp, -waypointSide
            elif value == 270:
                waypointUp, waypointSide = waypointSide, -waypointUp
        elif action == 'L':
            if value == 90:
                waypointUp, waypointSide = waypointSide, -waypointUp
            elif value == 180:
                waypointUp, waypointSide = -waypointUp, -waypointSide
            elif value == 270:
                waypointUp, waypointSide = -waypointSide, waypointUp

        elif action == 'F':
            shipUp += waypointUp * value
            shipSide += waypointSide * value
        # print(f'action={action} value={value}')
        # print(f'\tshipSide={shipSide} shipUp={shipUp}')
        # print(f'\twaypointSide={waypointSide} waypointUp={waypointUp}')

    return abs(shipUp) + abs(shipSide)


```

```python
part2(data)
```
