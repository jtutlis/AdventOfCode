```python
from timeit import default_timer as timer
from copy import deepcopy

f = open("data.txt", "r")
data = f.readlines()
data = [list(x.strip('\n')) for x in data]
```

```python
### Part 1
# Game of life? more like game of chairs

def inbounds(i,j,state):
    return 0 <= i < len(state) and  0 <= j < len(state[0])

def getCellsToCheck(i,j,state):
    cells = []
    for x in range(-1,2):
        for y in range(-1,2):
            # print(x,y, inbounds(i+x, j+y, length))
            if (x != 0 or y != 0) and inbounds(i+x, j+y, state):
                cells.append((i+x, j+y))
    return cells

def printState(state):
    for i in range(len(state)):
            for j in range(len(state[0])):
                print(state[i][j], end='')
            print()
    print()



def part1(chairs):
    # check if the state is the same as the previous state
    # update changes
    # save state
    listLen = len(chairs)

    prevState = None
    currState = deepcopy(chairs) ## Remains the same throughout updating state
    tempState = deepcopy(chairs) ## Gets updated throughout updating state 

    # printState(currState)
    count = 0

    while prevState != currState:
        prevState = deepcopy(currState)
        for i in range(len(chairs)):
            for j in range(len(chairs[0])):
                if currState[i][j] == 'L':
                    grow = True
                    for x,y in getCellsToCheck(i,j,currState):
                        # print(x,y)
                        if currState[x][y] == '#':
                            grow = False
                            break
                    if grow:
                        tempState[i][j] = '#'

                elif currState[i][j] == '#':
                    #If a seat is occupied (#) and four or more seats adjacent to it are also occupied, the seat becomes empty.
                    occupiedCount = 0
                    for x,y in getCellsToCheck(i,j,currState):
                        if currState[x][y] == '#':
                            occupiedCount += 1
                    if occupiedCount >= 4:
                        tempState[i][j] = 'L'
        currState = deepcopy(tempState)
        print(f'step={count}')
        # printState(currState)
        count +=1
        # input()
        
    occupiedCount = 0
    for i in range(len(chairs)):
            for j in range(len(chairs[0])):
                if currState[i][j] == '#':
                    occupiedCount += 1
    return occupiedCount
```

```python
### Part 2
# changes
# - people dont just care about adjacent seats, what they can see in a ray
# - takes 5 or more seats (rather than 4) for someone to leave a seat

def getRaysToCheck(i,j,state):
    # ooo
    # oxo
    # ooo
    rays = []
    deltas = []
    for x in range(-1,2):
        for y in range(-1,2):
            if x != 0 or y != 0:
                deltas.append((x,y))

    for x,y in deltas:
        ray = []
        currI = i + x
        currJ = j + y
        while inbounds(currI, currJ, state):
            ray.append((currI, currJ))
            currI += x
            currJ += y
        rays.append(ray)
    return rays

def part2(chairs):
    # check if the state is the same as the previous state
    # update changes
    # save state
    listLen = len(chairs)

    prevState = None
    currState = deepcopy(chairs) ## Remains the same throughout updating state
    tempState = deepcopy(chairs) ## Gets updated throughout updating state 

    count = 0

    while prevState != currState:
        prevState = deepcopy(currState)
        for i in range(len(chairs)):
            for j in range(len(chairs[0])):
                if currState[i][j] == 'L':
                    grow = True
                    for ray in  getRaysToCheck(i,j,currState):
                        for x,y in ray:
                            if currState[x][y] == '#':
                                grow = False
                                break
                            elif currState[x][y] == 'L':
                                break
                        if not grow:
                            break
                    if grow:
                        tempState[i][j] = '#'

                elif currState[i][j] == '#':
                    #If a seat is occupied (#) and four or more seats adjacent to it are also occupied, the seat becomes empty.
                    occupiedCount = 0
                    for ray in  getRaysToCheck(i,j,currState):
                        for x,y in ray:
                            if currState[x][y] == '#':
                                occupiedCount += 1
                                break
                            elif currState[x][y] == 'L':
                                break
                    if occupiedCount >= 5:
                        tempState[i][j] = 'L'
        currState = deepcopy(tempState)
        print(f'step={count}')

        count += 1
        
    occupiedCount = 0
    for i in range(len(chairs)):
            for j in range(len(chairs[0])):
                if currState[i][j] == '#':
                    occupiedCount += 1
    return occupiedCount
```
